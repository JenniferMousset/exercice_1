```python
import random  # import de la bibliothèque random afin d'utiliser choice pour choisir un mot aléatoire dans une liste
from math import ceil  # ceil va permettre de faire un arrondi par excès
import sys # import de la bibliothèque sys qui fournit un accès à certaines variables
import getpass # import de la bibliothèque getpass qui demande un mot de passe
import hashlib # import de la bibliothèque hashlib qui permet un hachage sécurisé du mot de passe
import csv # import de la bibliothèque csv qui implémente des classes pour lire et écrire des données tabulaires au format CSV



### Etape de Vérification ###
def Verification():
    """
        Name: Verification
        Objet: verifier si le fichier existe et s'il peut etre lu
        Output: affiche si le fichier n'existe pas ou s'il ne peut pas etre lu et appel la fonction Creation_compte
    """
    try :
        fichier = open("Dictionnaire_pendu.txt", "r")
        on_continu = True
    except IOError :
        print("Le fichier ne peut être trouvé.")
        on_continu = False
    if on_continu :
        if (fichier.readable()):
            on_continu = True
            Creation_compte()
        else :
            on_continu = False
            print("Le fichier ne peut être lu.")



def Creation_compte ():
    """
    Name : Creation_compte
    Objet : Créer un compte utilisateur pour permettre de jouer
    Output : Appel la fonction Authentification
    """
    nom_utilisateur1 = input("Choisissez un nom d'utilisateur : ")
    mdp = getpass.getpass ("Choisissez votre mot de passe : ").encode('utf8') # Affiche une demande de mot de passe sans renvoyer d'écho
    mdpc = hashlib.sha224(mdp).hexdigest() # Renvoie une chaîne de caractère contenant seulement des chiffres hexadécimaux
    ok = input ("Ecrivez ok en toutes lettres : ")
    if ok != "ok":
        print("Veuillez saisir ok")
        Creation_compte()
    refresh = "yes"
    compteur = 0
    Identification (nom_utilisateur1, mdp, mdpc, ok, refresh, compteur)

#### Enregistrement des données saisies dans un fichier csv ####
    with open('Stockage_données.csv', 'a', newline='') as csvfile:
        spamwriter = csv.writer(csvfile, delimiter='|',
                                quotechar='|', quoting=csv.QUOTE_MINIMAL)
        spamwriter.writerow([nom_utilisateur1] + [mdpc])


#### Identification de l'utilisateur ####
def Identification (nom_utilisateur1, mdp, mdpc, ok, refresh, compteur):
    """
        Name: Identification
        Objet: permet l'identification du joueur
        Output: verifie si le nom d'utilisateur et le mot de passe coincide avec la creation du compte
    """
    while ok == "ok" and refresh == "yes" :
        print("")
        print ("==== Identification ====")
        print("")
        nom_utilisateur2 = input ("Entrez votre nom d'utilisateur : ")
        mdp1 = getpass.getpass ("Tapez votre mot de passe : ").encode('utf8')
        mdpc2 = hashlib.sha224(mdp1).hexdigest()
        compteur = compteur + 1

        if nom_utilisateur2 == nom_utilisateur1 and mdpc2 != mdpc :
            print("")
            print ("Ce n'est pas le bon mot de passe !")

        if nom_utilisateur2 != nom_utilisateur1 and mdpc2 == mdpc :
            print("")
            print ("Le nom d'utilisateur que vous avez entré n'est pas valide.")

        if nom_utilisateur2 != nom_utilisateur1 and mdpc2 != mdpc :
            print("")
            print ("Les deux identifiants sont incorrects.")

        if compteur == 3 :
            print("")
            print ("Vous avez essayé", compteur, "fois... vous n'avez plus d'essais")
            Creation_compte ()

        if nom_utilisateur2 == nom_utilisateur1 and mdpc2 == mdpc :
            print("")
            print ("Vous avez mis les bons identifiants...")
            print("")
            print ("Bon jeu", nom_utilisateur2,"!")
            print("")
            print("=======================================")
            mots_aleatoires()
            break


### Choisi un mot aléatoire ###
def mots_aleatoires():
    """
    Name : mots_aléatoires
    Objet : Choisis un mot aléatoire dans le dictionnaire
    Return : retourne mot_aleatoire   
    """
    fichier = open("Dictionnaire_pendu.txt", "r")
    dictionnaire_pendu = list() #Création d'une liste nommée dictionnaire_pendu
    for mot in fichier: #Création d'une boucle pour récupérer tous les mots du fichier et les ajouter dans la liste dictionnaire_pendu (sans prendre les caractères non imprimables)
        dictionnaire_pendu.append(mot.rstrip())
    mot_aleatoire = random.choice(dictionnaire_pendu)#Création d'une variable mot_aleatoire qui prend un mot au hasard dans la liste
    return mot_aleatoire


### Afficher le nombre de lettre contenu dans le mot ###
def Affichage_du_nombre_de_lettre(Mot_choisi):
    """
    Name : Affichage_du_nombre_de_lettre
    Objet : Affiche le nombre de lettre contenu dans le mot
    Input : 1 arg de type str
    """
    print("Le mot possède:", len(Mot_choisi), "lettres")
    



def Affichage_sous_forme_tirets(Mot_choisi):
    """
    Name : Affichage_sous_forme_tirets
    Objet : Vérifie si un le mot contient un "-" il le place directement dans la liste et affiche le nombre de lettre contenu dans le mot sous forme de blanc souligné séparé par un espace
    Input : 1 arg de type str
    Return : list
    """
    Affiche_chaque_lettre_sous_forme_de_tiret = list()
    for lettre in range(len(Mot_choisi)):
      if "-" != Mot_choisi[lettre]:
        Affiche_chaque_lettre_sous_forme_de_tiret.append("_")  # Pour chaque lettre ajoute un "_" à la liste
      else :
        Affiche_chaque_lettre_sous_forme_de_tiret.append("-") # Place directement le "-" s'il est contenu dans le mot
    print(" ".join(Affiche_chaque_lettre_sous_forme_de_tiret))  # Affiche le contenu de la liste séparé par un espace
    return Affiche_chaque_lettre_sous_forme_de_tiret


def Nombre_de_tentative(Mot_affiche):
    """
    Name : Nombre_de_tentative
    Objet : Calcul le nombre de tentative pour deviner le mot, proportionnellement à la taille du mot. Effectue un arrondi par excès si le nombre divisé par 2 est un float.
    Input : 1 arg de type str
    Return : int
    """
    Nb_tentative = ceil(len(Mot_affiche) / 2)  # ceil arrondi par excès tel que 2.5 soit arrondi à 3
    print()
    print("Vous avez:", Nb_tentative, "tentatives")
    return Nb_tentative


### Rentrer une lettre ###
def rentrer(): ## A modifier "Rentrer"##
  """
  Name : rentrer
  Objet : Demande un input de la part du joueur et vérifie si l'entrée correspond bien à une lettre. 
  Return : str
  """
  print("")
  print("")
  lettre =input("Tapez une lettre : ")
  while not(lettre.isalpha()): # Si la lettre tapée n'est pas une lettre
    print("Merci de taper une lettre de l'alphabet.")
    lettre = input("Quelle est votre lettre : ")
    print("")
  return lettre

### La lettre n'est pas dans le mot ###
def rate(Mot_entre, Mot_defini, Tentative):
  """
  Name : rate
  Objet : Affiche raté si la lettre n'est pas dans le mot, calcul le nombre de tentatives restantes. Affiche aussi les lettres déjà tapées.
  Input : 3 args de type str, str et int
  Return : int, list
  """
  print("Raté")
  Tentative-=1 
  liste_rate.append(Mot_entre) # créé une liste avec les lettres rentrées
  print("")
  print("Les lettres que vous avez déjà tapées: ", " ".join(liste_rate))# Sépare chaque lettre de la liste avec des ","
  print("")
  return Tentative, liste_rate

    
### Si la lettre est dans le mot il affiche la lettre au bon emplacement de la liste de tiret###
def bravo(entre,mot):## A modifier "Bravo"   
  """
        Name: bravo
        Objet: permet d'afficher la liste de tiret contenant la bonne lettre à la bonne position.
        arg : 2 arguments de type str
        Output: affiche la liste de tiret avec la lettre
    """              
  for i in range(len(mot)):  
    if mot[i]==entre:
      Liste_sous_forme_de_tirets[i] = entre
      print("Bravo, la lettre se situe à la", i+1, "ème position ")
  print("")
  return Liste_sous_forme_de_tirets

### Echec quand il n'y a plus de tentative ###
def Echec (Tentative):
  """
        Name: Echec
        Objet: Vérifie si le nombre de tentative tombe à 0 et quitte le programme en affichant "Vous avez échoué!" si c'est le cas.
        arg : 1 arg de type int
    """  
  if Tentative == 0 :
    print("Vous avez échoué ! ")
    sys.exit(0)

### Réussite quand toutes les lettres ont été trouvé ###
def Reussite(Liste_sous_forme_de_tirets):
  """
  Name : Reussite
  Objet : Vérifie si toutes les lettres ont été trouvés et quitte le programme
  Input : 1arg de type list
  """
  a = str(Liste_sous_forme_de_tirets) # Cast de notre list sous forme d'un str afin d'utiliser la fonction find
  bravo_final = a.find("_")   # find va chercher dans notre mot s'il y a des blancs soulignés. Si il renvoi -1 alors ce veut dire qu'il ne trouve pas de blanc souligné.
  if bravo_final == -1:
    print("Bravo, vous avez trouvé le mot")
    sys.exit(0)


### 
def Boucle_principale(mot, Tentative, liste_rate, Liste_sous_forme_de_tirets):
  """
  Name : Boucle_principale
  Objet : boucle principale du programme qui appel les différentes fonctions afin de dérouler le jeu.
  Input : 4 args de type str, int, list, list
  """
  while True:
    entre = rentrer() 
    entre = entre.lower() 
    mot = mot.lower()   
    result = mot.find(entre) # Vérifie si la lettre est dans le mot. Renvoi -1 si la lettre n'est pas dans le mot.
    if result == -1:
      Tentative, liste_rate = rate(entre, mot, Tentative) # Récupère le retour de notre fonction afin de décrémenter nos tentatives et avoir notre liste de mot déjà utilisé
      print("Il vous reste: ",Tentative,"tentatives.")             
    else :
      Liste_sous_forme_de_tirets = bravo(entre, mot) # Récupère le retour de notre fonction afin d'avoir sous forme de liste les tirets remplacés par des lettres
      print(" ".join(Liste_sous_forme_de_tirets))
    Reussite (Liste_sous_forme_de_tirets)
    Echec (Tentative)


######################
# Fonction principal #                         
######################

Verification()

mot = mots_aleatoires()                     

Affichage_du_nombre_de_lettre(mot)             
Liste_sous_forme_de_tirets = Affichage_sous_forme_tirets(mot)      
Tentative = Nombre_de_tentative(mot)              

liste_rate = list() #ajout pour ne pas écraser notre liste à chaque fois
Boucle_principale (mot, Tentative, liste_rate, Liste_sous_forme_de_tirets)






```
