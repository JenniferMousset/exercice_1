```python
# #######################################
# JEU DE BATAILLE NAVALE A 1 BATEAU
# #######################################
# Objet : le programme suivant est un simple jeu de bataille navale où l'ordinateur place un bateau sur une matrice de 5x5.
# Version de Python : 3.8
# Bibliothèque(s) : random
# Environnement technique : Jupyter, Mac
# Auteur : 
# Mail : 
# Date de création : 
# Date de modification : 
# Version : 
# Licence : 
# #######################################

# La bibliothèque random sert à ce que le programme positionne au hasard le bateau.
import random

board = []

# Création d'une matrice de 5x5 qui correspond au plateau du jeu. Chaque case est représentée par un 0.
for x in range(0,5):
    board.append(["O"] * 5)

def print_board(board):
    ''' 
    Nom de la fonction : print_board
    Objet : ajouter un espace entre chaque case du plateau
    :Param : 1 argument de type list
    :Return : impression de la matrice bien formatée
    '''
    for row in board:
        print (" ".join(row))

print ("Let's play !")
print_board(board)

def random_row(board):
    ''' 
    Nom de la fonction : random_row
    Objet : choix au hasard de la ligne où est située le bateau
    :Param : 1 argument de type list
    :Return : int
    '''
    return random.randint(0,len(board)-1)

def random_col(board):
    ''' 
    Nom de la fonction : random_col
    Objet : choix au hasard de la colonne où est située le bateau
    :Param : 1 argument de type list
    :Return : int
    '''
    return random.randint(0,len(board[0])-1)

# #######################################
# PROGRAMME PRINCIPAL
# #######################################
ship_row = random_row(board)
ship_col = random_col(board)
print (ship_row)
print (ship_col)

# #######################################
# Le joueur doit trouver l'emplacement du bateau en 4 coups maximum.
# Selon les coordonnées qu'il donne, le programme affiche 4 réponses : 
#            - le joueur a trouvé les bonnées coordonnées => il gagne
#            - le joueur n'a pas trouvé au bout des 4 coups => le jeu se termine et les coordonnées s'affichent
#            - le joueur a tapé des coordonnées qui sortent du plateau => le programme lui indique que ces coordonnées ne sont pas bonnes
#.           - le joueur a déjà tapé les coordonnées => le programme l'avertit
#            - le joueur a tapé de bonnes coordonnées mais où le bateau n'est pas => le programme lui dit que c'est manqué
# #######################################
for turn in range(4):
    guess_row = input("Row:")
    guess_col = input("Col:")
    
    if int(guess_row) == ship_row and int(guess_col) == ship_col:
        print ("Congratulations! You win !")
        break
    else:
        if turn == 3:
            board[guess_row][guess_col] = "X"
            print_board(board)
            print ("Game Over")
            print ("I was here: [" + str(ship_row) + "][" + str(ship_col) + "]")
        else:
            if (int(guess_row) < 0 or int(guess_row) > 4) or (int(guess_col) < 0 or int(guess_col) > 4):
                print ("Oops, that's not even.")
            elif(board[int(guess_row)][int(guess_col)] == "X"):
                print ("You guessed that one already.")
            else:
                print ("You missed !")
                board[int(guess_row)][int(guess_col)] = "X" # Le signe X indique la position tapée par le joueur
            print (turn + 1)
            print_board(board)
```
