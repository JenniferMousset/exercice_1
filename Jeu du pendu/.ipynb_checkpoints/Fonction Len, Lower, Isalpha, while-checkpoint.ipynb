```python
# Nos importons la bibliothèque random. Cette bibliothèque contient des méthodes pour choisir
# de façon aléatoire un terme.
import random 

try :
    fichier = open("mots.txt", "r")
    print("Le fichier peut bien être chargé.")
    on_continu = True
except IOError :
    print("Le fichier ne peut être trouvé.")
    on_continu = False

if on_continu :
    if (fichier.readable()):
        print("Le fichier peut être lu.")
    else :
        print("Le fichier ne peut être lu.")

mots_pendu = [mot.rstrip() for mot in fichier]
print(mots_pendu)

mot_a_deviner = random.choice(mots_pendu)
print(mot_a_deviner)

print("Le mot contient ", len(mot_a_deviner), "lettres.")
for tiret in mot_a_deviner:
    print("- ", end='')
    
print()
lettre = input("Quelle est votre lettre : ")
if lettre.lower().isalpha(): # 2 methodes, lower et isalpha concatener # lettre est donc convertit en minuscule et on verifie si c'est bien une lettre
                             # possibilité de mettre un or après isalpha pour exclure une condition
    print("C'est bien une lettre.")
else : 
    print("Merci de taper une lettre de l'alphabet.")

if lettre.lower() in mot_a_deviner : 
    print("La lettre est bien dans le mot.")
    for lettre_mot in mot_a_deviner : 
        if lettre_mot == lettre.lower():
            print(lettre_mot, "", end='')
        else : 
            print("- ", end="")
else : 
    print("La lettre n'est pas dans le mot.")
```

```python
# Nos importons la bibliothèque random. Cette bibliothèque contient des méthodes pour choisir
# de façon aléatoire un terme.
import random 
try :
    fichier = open("mots.txt", "r")
    print("Le fichier peut bien être chargé.")
    on_continu = True
except IOError :
    print("Le fichier ne peut être trouvé.")
    on_continu = False
if on_continu :
    if (fichier.readable()):
        print("Le fichier peut être lu.")
    else :
        print("Le fichier ne peut être lu.")
mots_pendu = [mot.rstrip() for mot in fichier]
print(mots_pendu)
mot_a_deviner = random.choice(mots_pendu)
print(mot_a_deviner)
print("Le mot contient ", len(mot_a_deviner), "lettres.")
for tiret in mot_a_deviner:
    print("- ", end='')
    
print()
lettre = input("Quelle est votre lettre : ")
while not(lettre.lower().isalpha()): # tant que la personne n'a pas tapé une lettre alphabetique
    print("Merci de taper une lettre de l'alphabet.")
    lettre = input("Quelle est votre lettre : ")
else : 
    print("C'est bien une lettre.")
```
