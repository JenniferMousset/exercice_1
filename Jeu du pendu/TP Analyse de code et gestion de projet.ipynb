```python
# #######################################
# JEU DU MORPION
# #######################################
# Objet : le programme suivant est un jeu du morpion où l'ordinateur crée un plateau de jeu de 3x3 et ou...
# ... les 2 joueurs joueront chacun leur tour.
# Version de Python : 3.8
# Bibliothèque(s) : sys
# Environnement technique : Jupyter, Windows
# Auteur : Jennifer Mousset
# Mail : jenni.mousset@gmail.com
# Date de création : 17/04/2020
# Date de modification : 17/04/2020
# Version : 1.1
# Licence : 
# #######################################

# #######################################
# Built in function : print, int, input, 
# Variables :
# - board : variable de type de list pour créer le plateau 
# - player : variable de type str pour identifier le joueur
# - check_alignments : variable de type list pour concaténer les données X et O des joueurs
# - line : variable de type int pour indiquer la ligne sur le plateau
# - column : variable de type int pour indiquer la ligne sur le plateau
# - count_empty_cells : variable de type list pour compter les vides du tableau 
# Types de structures :
# - structure conditionnelle : if, else
# - structure itérative : while, for
# - Méthodes : sys
# - Bibliothèque sys : exit
########################################

# La bibliothèque sys .
import sys

# Création d'un plateau en 3x3 qui correspond au plateau de jeu. Initialement toutes les cases sont vides. On fait des tours 
# de jeu tant que le jeu ne s'arrête pas sur une égalité ou une victoire. Un tour de jeu consiste à faire jouer à un joueur
# des X et à l'autre joueur des O.
def game ():
    board = [
        [" ", " ", " "],
        [" ", " ", " "],
        [" ", " ", " "]
    ]
    while True:
        turn (board)
    ''' 
    Nom de la fonction : game
    Objet : creer un plateau de jeu 3x3 vide 
    :Param : 1 argument de type list
    :Instruction : faire des tours tant que le jeu ne s'arrête pas ou qu'il n'y a pas d'égalité 
    '''
def turn (board):
    play ("X", board)
    play ("O", board)
    ''' 
    Nom de la fonction : turn
    Objet : définir un tour de jeu
    :Param : 1 argument de type list
    :Return : int
    '''
    
def play (player, board):
    print_board (board)
    check_for_tie (board)
    print ("C'est au tour du joueur " + player + ".")
    print ("Ligne? (0, 1 ou 2)")
    line = int (input ())
    print ("Colonne? (0, 1 ou 2)")
    column = int (input ())
    if board[line][column] == " ":
        board[line][column] = player
        check_if_game_is_won (player, board)
    else:
        invalid_choice ()
    ''' 
    Nom de la fonction : play
    Objet : définir le joueur et le plateau / afficher le joueur et la ligne choisie
    :Param : 2 arguments de type list
    :Return : affiche le plateau de jeu et vérifie si possibilité de jouer. 
    :Intructions : si oui on modifie le plateau pour y mettre le symbole du joueur. si case non vide, stopper le jeu
    '''
def print_board (board):
    print ("-------")
    for i in range (0, 3):
        for j in range (0, 3):
            print ("|" + board[i][j], end="")
        print ("|")
        print ("-------")
    ''' 
    Nom de la fonction : print_board
    Objet : afficher les lignes de 0 à 2 puis une séparation de type | pour chaque case du plateau 
    :Param : 1 argument de type list
    :Return : affiche le plateau de jeu vide en 3x3 avec séparation |
    '''
def check_if_game_is_won (player, board):
    check_alignments (player, lines (board) + columns (board) + diagonals (board))
    ''' 
    Nom de la fonction : check_if_game_is_won
    Objet : concatener dans une meme liste les lignes, colonnes et diagonales
    :Param : 2 arguments de type list
    :Return : 
    '''

def check_alignment_for (player, line):
    if line == player * 3:
        game_is_over (player)
    ''' 
    Nom de la fonction : check_alignment_for
    Objet : définir ce qu'est un alignement
    :Param : 2 arguments de type list 
    '''
def check_alignments (player, candidates):
    for i in range (0, len (candidates)):
        check_alignment_for (player, candidates[i])
    ''' 
    Nom de la fonction : check_alignment
    Objet : tester tous les alignements possibles 
    :Param : 2 arguments de type list 
    '''
def columns (board):
    tboard = [[]] * 3
    for i in range (0, 3):
        tboard[i] = [" "] * 3
    for i in range (0, 3):
        for j in range (0, 3):
            tboard[j][i] = board[i][j]
    return tboard
    ''' 
    Nom de la fonction : columns
    Objet : renvoie les 3 colonnes du plateau
    :Param : 1 argument de type list
    :Return : les colonnes du plateau 
    '''
def lines (board):
    return (board)
    ''' 
    Nom de la fonction : lines
    Objet : renvoie les lignes du plateau
    :Param : 1 argument de type list
    :Return : les lignes du plateau 
    '''

def diagonals (board):
    return [
        [ board[0][0], board[1][1], board[2][2] ],
        [ board[2][0], board[1][1], board[0][2] ]
    ]
    ''' 
    Nom de la fonction : diagonals
    Objet : renvoie les diagonales du plateau
    :Param : 1 argument de type list
    :Return : les diagonales du plateau 
    '''
def count_empty_cells (board):
    c = 0
    for i in range (0, 3):
        for j in range (0, 3):
            if board[i][j] == " ":
                c = c + 1
    return c
    '''
    Nom de la fonction : count_empty_cells
    Objet : compte le nombre de case vide dans le plateau 
    :Param : 1 argument de type list
    :Return : X ou O  
    '''
def check_for_tie (board):
    if count_empty_cells (board) == 0:
        print ("Egalité!")
        sys.exit (0)
    '''
    Nom de la fonction : check_for_tie
    Objet : tester si il reste des cases vides sur le plateau 
    :Param : 1 argument de type list
    :Return : Egalité si plus de case vide 
    '''
def invalid_choice ():
    print ("Coup invalide!")
    sys.exit (0)
    '''
    Nom de la fonction : invalid_choice
    Objet : tester si le coup est valide, case vide ou non
    :Param : 
    :Return : Coup invalide puis stoppe le jeu
    '''
def game_is_over (player, board):
    print (player + " a gagné!")
    print_board (board)
    sys.exit (0)
    '''
    Nom de la fonction : game_is_over
    Objet : annoncer le vainqueur
    :Param : 2 arguments de type list
    :Return : a gagné! et stoppe le programme
    '''
# #######################################
# PROGRAMME PRINCIPAL
# #######################################
game ()
print_board (board)
```
