```python
import sys # import de la bibliothèque sys qui fournit un accès à certaines variables
import getpass # import de la bibliothèque getpass qui demande un mot de passe
import hashlib # import de la bibliothèque hashlib qui permet un hachage sécurisé du mot de passe
import csv # import de la bibliothèque csv qui implémente des classes pour lire et écrire des données tabulaires au format CSV
import pickle
#### Création compte utilisateur ####

def Creation_compte ():
    """
    Name : Creation_compte
    Objet : Créer un compte utilisateur pour permettre de jouer
    Input : 2 arg de type str
    Output : int
    """
    nom_utilisateur1 = input("Choisissez un nom d'utilisateur : ")
    mdp = getpass.getpass ("Choisissez votre mot de passe : ").encode('utf8') # Affiche une demande de mot de passe sans renvoyer d'écho
    mdpc = hashlib.sha224(mdp).hexdigest() # Renvoie une chaîne de caractère contenant seulement des chiffres hexadécimaux
    ok = input ("Ecrivez ok en toutes lettres : ")
    if ok != "ok":
        print("Veuillez saisir ok")
        Creation_compte()
    refresh = "yes"
    compteur = 0
    Identification (nom_utilisateur1, mdp, mdpc, ok, refresh, compteur)   

    
#### Enregistrement des données saisies dans un fichier csv ####


    with open('Stockage_données.csv', 'a', newline='') as csvfile:
        spamwriter = csv.writer(csvfile, delimiter='|',
                                quotechar='|', quoting=csv.QUOTE_MINIMAL)
        spamwriter.writerow(["nom du joueur"] + ["mot de passe"] + ["score"] + ["score global"])
        spamwriter.writerow([nom_utilisateur1] + [mdpc])



#### Identification de l'utilisateur ####

def Identification (nom_utilisateur1, mdp, mdpc, ok, refresh, compteur):
    while ok == "ok" and refresh == "yes" :
        print("")
        print ("==== Identification ====")
        print("")     
        nom_utilisateur2 = input ("Entrez votre nom d'utilisateur : ")
        mdp1 = getpass.getpass ("Tapez votre mot de passe : ").encode('utf8')
        mdpc2 = hashlib.sha224(mdp1).hexdigest()
        compteur = compteur + 1
        if nom_utilisateur2 == nom_utilisateur1 and mdpc2 != mdpc :
            print("")
            print ("Ce n'est pas le bon mot de passe !")
     
        if nom_utilisateur2 != nom_utilisateur1 and mdpc2 == mdpc :
            print("")
            print ("Le nom d'utilisateur que vous avez entré n'est pas valide.")
 
        if nom_utilisateur2 != nom_utilisateur1 and mdpc2 != mdpc :
            print("")
            print ("Les deux identifiants sont incorrects.")
     
        if compteur == 3 :
            print("")
            print ("Vous avez essayé", compteur, "fois... vous n'avez plus d'essais")
            Creation_compte ()
        
        if nom_utilisateur2 == nom_utilisateur1 and mdpc2 == mdpc :
            print("")
            print ("Vous avez mis les bons identifiants...")
            print("")
            print ("Bon jeu", nom_utilisateur2,"!")
        break
Creation_compte ()
```
