```python
# import de la bibliothèque random qui permet de tirer au hasard un réel compris dans un interval donné
import random
# créer un plateau
board = []

# boucle de 0 à 4
for x in range(0,5): 
  board.append(["O"] * 5) 
# append est une méthode qui ajoute les éléments dans une liste mais conserve la forme d'itérable. 
# La valeur de x est ['O','O','O','O','O'] puis il boucle 5 fois. 

def print_board(board):
  for row in board: 
    print " ".join(row) 

''' fonction print_board permet d'afficher un plateau 
:Param : l'argument de type string 
: Output : affiche tous les éléments du plateau séparé par un espace '''

print "Let's play !" # affiche le texte
print_board(board) # appelle de la fonction print_board avec la variable board

def random_row(board):
  return random.randint(0,len(board)-1)

''' fonction aléatoire renvoie un entier entre 0 et nombre lignes du plateau -1'''

def random_col(board):
  return random.randint(0,len(board[0])-1)

''' fonction aléatoire renvoie un entier entre 0 et nombre de colonnes contenu dans le plateau -1'''

ship_row = random_row(board) # affecte la fonction random_row(board) à la variable ship_row
ship_col = random_col(board) # affecte la fonction random_col(board) à la variable ship_col
print ship_row # affiche la variable ship_row
print ship_col # affiche la variable ship_col

for turn in range(4):
	guess_row = input("Row:")
	guess_col = input("Col:")
	# on va saisir un entier pour la variable guess_row à la question "Row:"
    # on va saisir un entier pour la variable guess_col à la question "Col:"
	if guess_row == ship_row and guess_col == ship_col:
    # si ce qui est saisi correspond exactement aux variables ship_row et ship_col définient préalablement on affiche 
     # "Congratulations! You win !" et on sort de la boucle.
	  print "Congratulations! You win !" 
	  break
    # sinon si on arrive au troisième tour le plateau est rempli de X. on affiche game over.on affiche la position.
	else:
		if turn == 3:
			board[guess_row][guess_col] = "X"
			print_board(board)
			print "Game Over"
			print "I was here: [" + str(ship_row) + "][" + str(ship_col) + "]"
            # sinon : 
            #si le chiffre saisi n'est pas compris entre 0 et 4 on affiche "Oops, that's not even."
            # si la position saisi a déjà été trouvé on affiche "You guessed that one already."
            # sinon on affiche "You missed !" 
            # A chaque erreur une croix se met dans le tableau dans la position saisie
		else:			
			if (guess_row < 0 or guess_row > 4) or (guess_col < 0 or guess_col > 4):
				print "Oops, that's not even."
			elif(board[guess_row][guess_col] == "X"):
				print "You guessed that one already."
			else:
				print "You missed !"
				board[guess_row][guess_col] = "X"
                # recommence la boucle for
			print (turn + 1) # affiche le tour actuel incrémenté de 1 à chaque fois 
			print_board(board) # affiche le plateau
```
